```python
#Mobile Recharge Validation System
def validate(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount >= 50 and amount in valid_plans:
        return True
    else:
        return False


# Loop for retry
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate(amount):
        print(" Recharge Successful!")
        break
    else:
        print(" Invalid recharge. Try again.\n")
```

```python
#Inventory Reorder Alert System
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, " Reorder Alert ")
        else:
            print(item, " Stock OK ")


# Given dictionary
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

# Call function
check_inventory(inventory)
```

```python
#Student Result Processing System
def process_result(marks):
    total = 0
    
    # Loop to calculate total
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Example marks list
student_marks = [60, 45, 70, 55]

avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

```python
#Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak == "yes":
        fare = fare + (fare * 0.25)   # 25% extra
    
    return fare


# Retry system using while loop
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ").lower()
        
        total_fare = calculate_fare(km, peak_hour)
        
        print("Total Fare: ₹", total_fare)
        
        again = input("Do you want to calculate again? (yes/no): ").lower()
        if again != "yes":
            print("Thank you for using Cab Fare Estimator ")
            break
            
    except ValueError:
        print(" Please enter valid distance.\n")
```

```python
#Employee Attendance Eligibility 
def check_eligibility(attendance):
    present_days = 0
    
    # Count Present days
    for day in attendance:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance)) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_eligibility(attendance_list)

print("Attendance Percentage:", percent)
print("Status:", status)
```

```python
#Password Strength Checker
def check_password(pwd):
    if len(pwd) < 8:
        return "Weak Password"

    digit = False
    special = False

    for ch in pwd:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True

    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(check_password(password))
```

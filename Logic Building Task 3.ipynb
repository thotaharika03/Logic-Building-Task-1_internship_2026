```python
#1.Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#  Find the highest score
highest_score = max(employees.values())

# Find employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

```python
#2.Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

# Convert to lowercase and split into words
words = query.lower().split()

frequency = {}

# Count words
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Print only words repeated more than once
result = {}

for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)
```

```python
# 3.Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through list with index
for index in range(len(sensor_readings)):
    if sensor_readings[index] % 2 == 0:   # Check if even
        valid_readings.append((index, sensor_readings[index]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4. Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

for domain, count in domain_count.items():
    print(f"{domain}: {int((count/total)*100)}%")


```

```python
#5.Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average = sum(sales) / len(sales)

# Step 2: Detect spikes (more than 30% above average)
threshold = average * 1.3

for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")

```

```python
#6. Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Step 1: Count occurrences of each user ID
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Step 2: Display only duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} -> {count} times")
```

```python
#Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_students = 0
fail_students = 0
for mark in marks:               #for loop to check each mark.
    if mark >= 50:         
        pass_students += 1
    else:
        fail_students += 1

print("Total Students:", len(marks))
print("Number of Students Passed:", pass_students)
print("Number of Students Failed:", fail_students)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for i in names:
    cleaned=i.strip().lower()     #strip(): removes extra spaces from beginning and end,lower(): converts text to lowercase
    cleaned_names.append(cleaned)
print(cleaned_names)              #Printing inside the loop prints each name
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)          #len(message):calculates number of characters
    print("Message:", msg)
    print("Length:", length)
    
    if length > 10:                     #length > 10 : flag as long message
        print("Status: Long Message ")
    else:
        print("Status: Short Message")  
```

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
Error_count=0
for log in logs:
    if log == "ERROR":            #if log == "ERROR" : only then increase counter
        Error_count+=1        
print("error_count:",Error_count)
```

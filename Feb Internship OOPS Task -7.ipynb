```python
#1.Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default status

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Sample Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

# Creating object
light = SmartLight(light_name)

# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")

# Display Output
light.display_status()
```

```python
#2.Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating object
employee = Employee(name, emp_id, department)

# Displaying ID Card
employee.display_id_card()
```

```python
#3.Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Taking input
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Creating object
contact = Contact(name, phone)

# Displaying contact information
contact.display_contact()
```

```python
#4.Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# Taking input
name = input("Product Name: ")
price = input("Price: ")

# Creating object
product = Product(name, price)

# Displaying price tag
product.display_price_tag()
```

```python
#5.Product Price Tag Generator
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# Taking input
name = input("Movie Name: ")
rating = input("Rating: ")

# Creating object
movie = Movie(name, rating)

# Displaying movie details
movie.display_details()
```

```python
#6.Delivery Address Manager
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Taking input
name = input("Customer Name: ")
address = input("Address: ")

# Creating object
delivery = Delivery(name, address)

# Displaying delivery details
delivery.display_delivery_details()
```

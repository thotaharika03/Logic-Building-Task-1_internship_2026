```python
#1.Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes=0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    # Condition to check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes = [200, 300, 150, 400]   # Total = 1050
analyze_engagement(likes)
    
```

```python
#2.Healthcare – Medicine Stock Alert System

def medicine_stock(stock_items):
    total_stock = sum(stock_items)
    
    print("Medicine Stock:", total_stock)
    
    if total_stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


stocks = [1, 6, 5, 5, 4, 3, 9, 6]

medicine_stock(stocks)
        
```

```python
#3.Agriculture – Rainfall Adequacy Checker
# Seasonal Rainfall Analysis System

def rainfall_analysis(rainfall_data, required_level):
    total = 0
    
    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total += rain
    
    # Calculate average rainfall
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", int(average))
    
    # Check rainfall status
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example Data
seasonal_rainfall = [70, 75, 68, 80, 67]
required_level = 70

rainfall_analysis(seasonal_rainfall, required_level)
```

```python
#4.Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set()
    duplicate_found = False
    
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Data
users = ["user1", "user2", "user3", "user1"]

detect_duplicates(users)
```

```python
#6.Agriculture – Premium Crop Price Filter
def crop_prices(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    print("Premium Crops:",premium)
price=[2000,2500,1500,3200]
crop_prices(price)
```

```python
#7.System Monitoring – Application Health 
def health_based(error):
    error_count=0
    for errors in error:
        error_count+=errors
    print("Error_Count:",error_count)
    
    if error_count==0:
        print("system status:Healthy")
    elif error_count <=5:
        print("system status:Minor Issues")
    else:
        print("system status:Critical Issues")
errors=[2,3,2]
health_based(errors)
```

```python
#9.E-Learning – Student Attendance Eligibility System
def attendance_percentage(attendance):
    percentage=(sum(attendance)/len(attendance))*100
    print("Attandance Percentage:",percentage)
    if percentage>=75:
        print("Exam Eligibility:Eligible")
    else:
        print("Exam Eligibility:Not Eligible")
attendance_data=[1,1,1,1,0] # 4 out of 5 days present → 80%(1 = Present, 0 = Absent)
attendance_percentage(attendance_data)
```

```python
#10Smart Electricity Bill Analyzer
def calculate_bill(units):
    # Billing calculation using slabs
    if units<=100:
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:",status)
calculate_bill(250)
```
